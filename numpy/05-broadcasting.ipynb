```python
import numpy as np
```

```python
# Broadcasting
# rules: values has to have at least 1 equal dim. length or be 1 dim.

a = np.array([0, 1, 2])
b = np.array([5])
a + 2 # returns an array with each element of "a" + 2
```

```python
# rows
M = np.ones((2, 3))
a = np.arange(3)
a, a.shape, M, M.shape, a + M
```

```python
# cols
M = np.ones((3, 2))
a = a.reshape((3, 1))
a, a.shape, M, M.shape, a + M
```

```python
# error
a = np.arange(4) # [0, 1, 2, 3] shape: (4, )
M = np.ones((2, 2)) # [[1, 1], [1, 1]] shape: (2, 2)
try:
    a + M # Error
except ValueError as e:
    print(e)
```

```python
a = np.arange(3) # [0, 1, 2] shape: (3, )
M = np.ones((3, 2)) # [[1, 1, 1], [1, 1, 1] shape: (3, 2)
try:
    a + M # Error
except ValueError as e:
    print(e, a, M, sep='\n')
```
